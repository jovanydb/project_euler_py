```python
import math


def solve15(width: int, length: int) -> int:
    """
    Compute the number of lattice paths through a grid of size (width * length),
    moving only right or down.

    Formula:
        C(width + length, width) = (width + length)! / (width! * length!)
    """
    return math.comb(width + length, width)


if __name__ == "__main__":
    print(solve15(20, 20))
```
