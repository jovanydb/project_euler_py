```python
def solve2():
    fib_nums = [1, 2]

    # Generate Fibonacci numbers below 4 million
    while fib_nums[-1] < 4_000_000:
        fib_nums.append(fib_nums[-2] + fib_nums[-1])

    # Exclude the last number if it exceeds the limit
    fib_nums.pop()

    # Sum all even Fibonacci numbers
    return sum(num for num in fib_nums if num % 2 == 0)


if __name__ == "__main__":
    print(solve2())
```
