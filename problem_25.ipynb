```python
def solve25() -> int:
    """
    Return the index of the first Fibonacci number that has 1000 digits.
    Iterative O(n) time, O(1) space.
    """
    a, b = 1, 1  # F1, F2
    index = 3    # we'll compute c = F3 first

    while True:
        c = a + b
        a, b = b, c

        # Check digit length by converting to string (simple and reliable)
        if len(str(c)) == 1000:
            break

        index += 1

    return index


if __name__ == "__main__":
    print(solve25())
```
