```python
def Solve25():
    # generate Fibonacci numbers iteratively, not recursively. O(n)?, O(1)
    a = 1
    b = 1
    index = 3
    while True:
        c = a + b
        a = b
        b = c
        # turn them into strings and check length of given string. Still should be O(n)? and O(1)
        if len(str(c)) == 1000:
            break
        index += 1
    return index

print(Solve25())
```
