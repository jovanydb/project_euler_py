```python
def solve10(limit: int) -> int:
    """
    Return the sum of all prime numbers below `limit`.

    This implementation incrementally builds a list of primes and
    tests each odd number for divisibility up to its square root.
    """
    if limit < 2:
        return 0

    primes = [2, 3, 5]
    prime_sum = sum(primes)
    candidate = 7

    while candidate <= limit:
        is_prime = True

        for prime in primes:
            # Every composite number has a prime factor <= sqrt(candidate)
            if prime * prime > candidate:
                break
            if candidate % prime == 0:
                is_prime = False
                break

        if is_prime:
            primes.append(candidate)
            prime_sum += candidate

        candidate += 2  # skip even numbers

    return prime_sum


if __name__ == "__main__":
    print(solve10(2000000))
```
