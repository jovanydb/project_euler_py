```python
def Solve10(x):
    # setting up data structure for storing prime numbers
    primeList = [2, 3, 5]
    primeSum = sum(primeList)
    counter = 7

    while counter <= x:
        is_prime = True
        for prime in primeList:
            # every composite number have at least one prime factor less than its square root
            if prime * prime > counter:
                break
            # checks the list of primes and "crosses-out" those which are multiples of that prime
            if counter % prime == 0:
                is_prime = False
                break
        # if the number is prime it appends it and continues the loop skipping over even numbers
        if is_prime:
            primeList.append(counter)
            primeSum += counter
        counter += 2
    
    return primeSum

print(Solve10(2000000))
```
