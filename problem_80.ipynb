```python
def decimalSqrt(number: int, precision: int = 120, convergence: int = 30):
    from decimal import Decimal, getcontext
    # add a buffer for intermediate steps
    x = Decimal(number)

    for _ in range(convergence):
        x_prev = x
        x = (x + Decimal(number) / x) / 2
        # checks for early convergence
        if x == x_prev:
            break
    getcontext().prec = precision
    return x

def rounder(givenList):
    for items in givenList:
            if items[101] >= 5:
                items[100] += 1
    return givenList

def Solve80(precision: int):
    import math
    # begin by finding all natural numbers that have irrational square roots
    iRoots = [i for i in range(1, 101) if int(math.sqrt(i)) != math.sqrt(i)]
    iRoots = [str(decimalSqrt(num)).split(".")[1] for num in iRoots]
    listifiedDigits = [list(digits) for digits in iRoots]
    intifiedListOfDigits = []
    for items in listifiedDigits:
        digits = []
        intifiedListOfDigits.append(digits)
        for digit in items:
            digits.append(int(digit))
    updated_list = rounder(intifiedListOfDigits)
    updated_list = [items[:len(items)-1] for items in updated_list]
    res = sum(map(sum, updated_list))
    return res

print(Solve80())
```
