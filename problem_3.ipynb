```python
def largest_prime_factor(n: int) -> int | None:
    """Return the largest prime factor of n, or None if n < 2."""
    if n < 2:
        return None

    # Strip out factor 2
    last = None
    while n % 2 == 0:
        last = 2
        n //= 2

    # Check odd factors
    f = 3
    while f * f <= n:
        while n % f == 0:
            last = f
            n //= f
        f += 2

    # Whatever remains (>1) is prime and is the largest factor
    return n if n > 1 else last


def solve3(x: int) -> int | None:
    return largest_prime_factor(x)


if __name__ == "__main__":
    print(solve3(600851475143))
```
