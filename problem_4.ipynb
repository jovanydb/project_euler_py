```python
# My Solution
import math


def is_palindrome(x: int) -> bool:
    """Return True if x is a palindrome."""
    return str(x) == str(x)[::-1]


def find_factors(x: int) -> list[int]:
    """Return a list of factors of x (excluding 1 and x)."""
    return [i for i in range(2, int(math.sqrt(x)) + 1) if x % i == 0]


def solve4() -> int | None:
    """Find the largest palindrome made from the product of two 3-digit numbers (naive)."""
    nums = list(range(1, 999 * 999 + 1))
    palindromes = [n for n in nums if is_palindrome(n)]
    palindromes.reverse()

    for num in palindromes:
        for factor in find_factors(num):
            quotient = num / factor
            if 100 <= quotient <= 999:
                return num
    return None


# More efficient solution (5 to 10 times faster)
def solve4_efficient() -> int:
    """Find the largest palindrome made from the product of two 3-digit numbers (optimized)."""
    max_palindrome = 0

    for i in range(999, 99, -1):
        for j in range(i, 99, -1):
            product = i * j
            if product <= max_palindrome:
                break
            if is_palindrome(product) and product > max_palindrome:
                max_palindrome = product
    return max_palindrome


if __name__ == "__main__":
    print(solve4())
    print(solve4_efficient())
```
