```python
# My Solutiom
import math

def isPalindrome(x: int):
    if str(x) == str(x)[::-1]:
        return int(x)

def findFactors(x: int):
    factors = [i for i in range(2,int(math.sqrt(x))+1) if x%i == 0]
    return factors

def Solve4():
    nums = list(range(1,999*999+1))
    paliNums = list(filter(isPalindrome, nums))[::-1]
    paliFactors = [findFactors(i) for i in paliNums]
    
    for num, factors in zip(paliNums, paliFactors):
        for factor in factors:
            x = num / factor
        if 100 <= x <= 999:
            return num

# More efficient solution (5 to 10 times faster)

def Solve4_Efficient():
    max_palindrome = 0

    for i in range(999, 99, -1):
        for j in range(i, 99, -1):
            product = i * j
            if product <= max_palindrome:
                break 
            if isPalindrome(product):
                if product > max_palindrome:
                    max_palindrome = product
    return max_palindrome

print(Solve4())
print(Solve4_Efficient())
```
