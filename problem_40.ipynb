```python
def nthConcat(n: int):
    # given an int n, it concatenates all numbers within the range into a string
    digit = ""
    for i in range(n+1):
        digit += str(i)
    return digit

def Solve40(n):
    # nthConcat() returns a stringified number, which is represented by a digit in a list, which is then intified, summed and returned by the Solve40()
    champConst = list(nthConcat(1000000))
    res = 1
    for i in range(n):
        res *= int(champConst[10**i])
    return res

print(Solve40(7))
```
