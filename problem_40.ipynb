```python
def nth_concat(n: int) -> str:
    """Concatenate all integers from 0 to n into a single string."""
    return "".join(str(i) for i in range(n + 1))


def solve40(n: int) -> int:
    """
    Compute the product of the digits at positions 1, 10, 100, ..., 10^(n-1)
    in Champernowne's constant (concatenation of positive integers).
    """
    champ_const = nth_concat(1_000_000)
    result = 1

    for i in range(n):
        # subtract 1 because Python uses 0-based indexing
        result *= int(champ_const[10**i - 1])

    return result


if __name__ == "__main__":
    print(solve40(7))
```
