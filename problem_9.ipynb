```python
def generate_pythagorean_triplets(limit: int) -> list[list[int]]:
    """Generate all Pythagorean triplets (a, b, c) with a, b, c < limit."""
    triplets = []
    for a in range(1, limit):
        for b in range(a, limit):
            c = (a**2 + b**2) ** 0.5
            if c.is_integer():
                triplets.append([a, b, int(c)])
    return triplets


def solve9(limit: int) -> int | None:
    """Find the product a*b*c of the Pythagorean triplet where a + b + c = limit."""
    for a in range(1, limit):
        for b in range(a, limit):
            c = (a**2 + b**2) ** 0.5
            if c.is_integer() and a + b + c == limit:
                return int(a * b * c)
    return None


if __name__ == "__main__":
    print(solve9(1000))
```
