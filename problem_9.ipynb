```python
def GeneratePyTriplets(limit: int):
    triplets = []
    for a in range(1, limit):
        for b in range(a, limit):
            c = ((a**2 + b**2)**0.5)
            if c.is_integer():
                triplets.append([a,b,int(c)])  
    return triplets

def Solve9(limit: int):
    for a in range(1, limit):
        for b in range(a, limit):
            c = ((a**2 + b**2)**0.5)
            if c.is_integer() and a + b + c == limit:
                return int(a*b*c)

print(Solve9(1000))
```
