```python
def solve6(x: int) -> int:
    """Return the difference between the square of the sum and the sum of the squares of the first x natural numbers."""
    sum_squares = sum(i * i for i in range(1, x + 1))
    square_sum = sum(range(1, x + 1)) ** 2
    return square_sum - sum_squares


if __name__ == "__main__":
    print(solve6(100))
```
