```python
def factorial(n: int) -> int:
    """Compute n! recursively."""
    if n == 0:
        return 1
    return n * factorial(n - 1)


def solve20(n: int) -> int:
    """Return the sum of the digits of n!."""
    return sum(int(d) for d in str(factorial(n)))


if __name__ == "__main__":
    print(solve20(100))
```
