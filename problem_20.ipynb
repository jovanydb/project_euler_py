```python
# generates factorials recursively
def factorial(n: int):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# sums the digit of 100 factorial
def Solve20(n: int):
    digitNums = str(factorial(n))
    res = 0
    for digit in digitNums:
        res += int(digit)
    
    return res

print(Solve20(100))
```
