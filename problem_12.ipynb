```python
import numpy as np

def GenerateNTriNums(n: int):
    arr = np.arange(1, n + 1, dtype = np.int32)
    trinums = arr * (arr + 1) // 2
    return trinums

def Solve12(n: int):
    trinums = GenerateNTriNums(100000)
    primeNums = [2,3,5,7,11,13]

    for i in primeNums:
        trinums = trinums[trinums % i == 0]

    divisors = []
    res = 0

    for num in trinums:
        if len(divisors) > 500:
            break
        divisors = []
        for divisor in range(1, num + 1):
            if len(divisors) > 500:
                break
            if num%divisor == 0:
                divisors.append(divisor)
        res = num

    return res

print(Solve12(100000))
```
